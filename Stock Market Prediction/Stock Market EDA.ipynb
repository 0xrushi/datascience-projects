# Simple Moving Averages

A simple moving average (SMA) is an arithmetic moving average calculated by adding recent closing prices and then dividing that by the number of time periods in the calculation average.

```python
TSLA['Close'].loc['2018-01-01':'2019-01-01'].ta_plot(study='sma',periods=[13,21,55],title='Simple Moving Averages')
```

```python
WMT['Close'].loc['2018-01-01':'2019-01-01'].ta_plot(study='sma',periods=[13,21,55],title='Simple Moving Averages')
```

```python
XIACF['Close'].loc['2018-01-01':'2019-01-01'].ta_plot(study='sma',periods=[13,21,55],title='Simple Moving Averages')
```

# Bollinger Band Plot

A Bollinger Band is a technical analysis tool defined by a set of lines plotted two standard deviations (positively and negatively) away from a simple moving average (SMA) of the security's price, but can be adjusted to user preferences.

```python
TSLA['Close'].loc['2018-01-01':'2019-01-01'].ta_plot(study='boll')
```

```python
WMT['Close'].loc['2018-01-01':'2019-01-01'].ta_plot(study='boll')
```

```python
XIACF['Close'].loc['2018-01-01':'2019-01-01'].ta_plot(study='boll')
```

# Risk Calculation

```python
df.head()['Close']
```

```python
dfcomp = df['Adj Close']
```

```python
dfcomp.head()
```

```python
retscomp = dfcomp.pct_change()
corr = retscomp.corr()
corr
```

```python
plt.scatter(retscomp.TSLA, retscomp.WMT)
plt.xlabel('Returns TSLA')
plt.ylabel('Returns WMT')
```

```python
plt.scatter(retscomp.TSLA, retscomp.XIACF)
plt.xlabel('Returns TSLA')
plt.ylabel('Returns XIACF')
```

```python
plt.scatter(retscomp.XIACF, retscomp.WMT)
plt.xlabel('Returns XIACF')
plt.ylabel('Returns WMT')
```

```python
#RISK CALCULATION
plt.scatter(retscomp.mean(), retscomp.std())
plt.xlabel('Expected returns')
plt.ylabel('Risk')
for label, x, y in zip(retscomp.columns, retscomp.mean(), retscomp.std()):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (20, -20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```

Finally, the risk calculation clearly shows the risk to return ratio for our 3 stocks. We can safely say that picking 3 differently performing stocks, while Tesla & Walmart are relatively safer stocks to invest in, Xiaomi is a rather riskier stock to invest in. Reasons being, Xiaomi went public not so long ago & not performing well in the time we have performed our analysis on.
